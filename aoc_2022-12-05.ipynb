#### Clean each move instruction

```python
def clean_moves(step):
    clean_step = step.replace('move', '').replace('from', ',').replace('to',',').replace(' ','')
    return clean_step.split(',')

move_list= np.asarray([clean_moves(step) for step in moves], dtype = int)
move_list[0:5]
```

```python
def move_cargo( n, move_from, move_to, cargo_list = crate_list):

    new_cargo_list = cargo_list
    
    cargo_to_move = cargo_list[move_from-1][-n::]
   
    new_cargo_list[move_from-1]= cargo_list[move_from-1][0:-n]
    new_cargo_list[move_to-1] = np.concatenate(( cargo_list[move_to - 1], cargo_to_move[::-1]))
    
    return(new_cargo_list)


```

```python
def move_cargo_multi( n, move_from, move_to, cargo_list = crate_list):

    new_cargo_list = cargo_list
    
    cargo_to_move = cargo_list[move_from-1][-n::]
   # print('Cargo to move', cargo_to_move)

    new_cargo_list[move_from-1] = cargo_list[move_from-1][0:-n]
    
    new_cargo_list[move_to-1] = np.concatenate(( cargo_list[move_to - 1], cargo_to_move))
    
    return(new_cargo_list)

```
