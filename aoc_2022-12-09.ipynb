```python
import numpy as np

# Initialize Grid
grid = np.repeat(None, 60).reshape(2, 5, 6)
grid[0, 4, 0] = 'H'
grid[1, 4, 0] = 'T'

print(grid)




def move_knot(mat, end, direction ):
    
    # mat = matrix 
    # end = H/T Heads or Tails
    # Direction
    
    if direction == 'U':
        direct_xy = 1
        direct_sign = -1
    elif direction == 'D':
        direct_xy = 1
        direct_sign = 1
    elif direction == 'L':
        direct_xy = 2
        direct_sign = -1
    elif direction == 'R':
        direct_xy = 2
        direct_sign = 1
     
    # Find Current Position
    cur_pos= np.where(mat == end)

    # Set New Position for Heads
    cur_pos[direct_xy][0] = cur_pos[direct_xy][0] + direct_sign
          
    print(min(cur_pos))
            
    if (min(cur_pos) < 0) | (max(cur_pos) +1 > mat.shape[direct_xy]):
        print('Error: Out of Bounds')
        exit()
        
    # Moveaway from current position
    mat[cur_pos] = None
    
        
    # move Tails to New Position
    mat[cur_pos] = end
    
    
    return(mat)
    
    
move_knot(grid, 'H', 'L')
grid.shape  
```
