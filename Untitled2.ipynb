### Move Knot Function

```python

import numpy as np

# Initialize Grid
grid = np.repeat(None, 60).reshape(2, 5, 6)
grid[0, 4, 0] = 'H'
grid[1, 4, 0] = 'T'

print(grid)

def move_heads(mat, direction ):

    # mat = matrix/grid 
    
    # Determine x/y direction/sign based on direction input
    if direction == 'U':
        direct_xy = 1
        direct_sign = -1
    elif direction == 'D':
        direct_xy = 1
        direct_sign = 1
    elif direction == 'L':
        direct_xy = 2
        direct_sign = -1
    elif direction == 'R':
        direct_xy = 2
        direct_sign = 1
     
    # Find Current Position
    cur_pos= np.where(mat == 'H')
    
    # Calculate new (potential) position
    new_pos =  cur_pos[direct_xy][0] + direct_sign
    
          
    # Return Same Matrix if New Heads Position is Outof Pounds
    
    if (new_pos < 0) | (new_pos +1 > mat.shape[direct_xy]):
        print('Error: Out of Bounds')
        return(mat)
    
    else:

        # Moveaway from current position
        mat[cur_pos] = None

        # Set New Position for Heads
        cur_pos[direct_xy][0] = new_pos

        # move Tails to New Position
        mat[cur_pos] = 'H'


        return(mat)
    
    
move_heads(grid, 'R')
#grid.shape  
```

### move_tails( ) function

```python
# Initialize Grid
grid = np.repeat(None, 60).reshape(2, 5, 6)
grid[0, 4, 0] = 'H'
grid[1, 4, 0] = 'T'

# Definf follow_head

def move_tails(grid):
    
    # Compare Positions
    vert = np.where(grid[0] == 'H')[0][0] - np.where(grid[1] == 'T')[0][0]
    horiz =  np.where(grid[0] == 'H')[1][0] - np.where(grid[1] == 'T')[1][0]
    
    #print([horiz, vert])

 
    # Tails Moves Right & Up
    if ((horiz >= 1) & (vert <= - 2)) | ((horiz >= 2) & (vert <= -1)):
        
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] +  1
        cur_pos_T[1][0] = cur_pos_T[1][0] + -1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
        
    # Tails Moves Left and Down
    elif ((horiz <= -1) & (vert >= 2)) | ((horiz <= -2) & (vert >= 1)):
        
                
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] + -1
        cur_pos_T[1][0] = cur_pos_T[1][0] +  1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
        
    
        
    # Tails Moves Up and Left
    elif ((horiz <= -2) & (vert <= -1)) | ((horiz <= -1) & (vert <= -2)):
        
                
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] + -1
        cur_pos_T[1][0] = cur_pos_T[1][0] + -1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
        
    
    # Tails Moves Down & Right
    elif ((horiz >= 2) & (vert >= 1)) | ((horiz >= 1) & (vert >= 2)):
        
                
       # print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] + 1
        cur_pos_T[1][0] = cur_pos_T[1][0] + 1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
    
    
    # Tails Moves Down
    elif vert >= 2:
            
       # print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[1][0] = cur_pos_T[1][0] + -1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
    
        
    # Tails Moves Up
    elif vert <= -2:
            
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
        
        
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[1][0] = cur_pos_T[1][0] + -1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
    
        
    # Tails Moves Right 
    elif horiz >= 2:
        
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')
 
        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] + 1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
        
    # Talils Moves Left
    elif horiz <= -2:
        
        #print('Move')
        
        # Find Current Position of Tails
        cur_pos_T = np.where(grid == 'T')

        # Move Tails away from current position
        grid[cur_pos_T] = None
        
        # Set New Position for Tails
        cur_pos_T[2][0] = cur_pos_T[2][0] - 1
        
        # move Tails to New Position
        grid[cur_pos_T] = 'T'
        
    return(grid)
        
```
