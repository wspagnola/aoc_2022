```python
with open('data/input_2022-12-03.txt') as f:
    text = [x.strip() for x in f.readlines()]
    

sacks = []
for t in text:
    t_split = int(len(t) /2)
    
    sacks.append([set(t[0:t_split]), set(t[t_split::])])
  #  print(t)
  #  print(t[0:t_split])
  #  print(t[t_split::])
  #  print('\n')
    
sacks[0:6]
```
